```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
```

# About Columns 

## Distribution of Numeric Columns

```python
fig, axes = plt.subplots(2,3, figsize=(15,8))
df.hist(ax=axes)
[x.title.set_size(18) for x in axes.ravel()]
fig.suptitle('Distribution of Numeric Columns', size=25)
plt.show()
```

## Boxplot of Numeric Columns

```python
def adjust_boxplot():
    plt.box(False)
    plt.gca().yaxis.label.set_size(18)
    plt.gca().yaxis.labelpad = 0
    plt.tick_params(bottom=False, left=False, labelsize=13, pad=-55)
    plt.tight_layout()
    

plt.figure(figsize=(14,4))
plt.suptitle('Boxplot of Numeric Columns', fontsize=27)

num_cols = ['rating', 'salary_avg', 'description_len', 'age']

for i in range(len(num_cols)):
    plt.subplot(1, len(num_cols), i+1)
    sns.boxplot(y=df[num_cols[i]], width=0.2)
    adjust_boxplot()
```

## Correlation Heatmap of Numeric Columns

```python
plt.figure(figsize=(8,4))
cmap = sns.color_palette('coolwarm', as_cmap=True)
heatmap = sns.heatmap(df[num_cols].corr(), annot=True, fmt='.1g', vmin=-0.1, vmax=0.4, center=0, 
                      linewidths=1, linecolor='Black', cmap=cmap, cbar_kws={'shrink': 0.7})
heatmap.set_title('Correlation Heatmap of Numeric Columns', fontdict={'fontsize': 20}, pad=20)
heatmap.tick_params(labelsize=13)
plt.show()
```

## Distribution of Categorical Columns

```python
df_cat = df[['location', 'size', 'type_of_ownership', 'industry', 'sector', 'revenue', 'seniority']]

for i in df_cat.columns:
    order = df_cat[i].value_counts().sort_values(ascending=False).index
    plt.figure(figsize=(5,3))
    sns.countplot(x=df_cat[i], order=order)
    plt.gca().set_frame_on(False)
    plt.gca().xaxis.set_label_position("top")
    plt.gca().xaxis.label.set_size(18)
    plt.ylabel('')
    plt.tick_params(labelrotation=90)
    plt.show() 
```

# Questions

## Which Position Has the Highest Salary in Average? 

```python
def adjust_plot(title=''):
    plt.title(title, fontdict={'fontsize': 18}, pad=20)
    #plt.gca().xaxis.set_label_position("top")
    plt.gca().xaxis.label.set_size(15)
    plt.gca().xaxis.set_label_coords(0.5,-0.1)
    plt.ylabel('')
    plt.show()
```

```python
pt1 = pd.pivot_table(df, index='job_title', values='salary_avg').sort_values(by='salary_avg', ascending=False)

sns.barplot(pt1, x='salary_avg', y=pt1.index, orient='h')
adjust_plot(title='Average Salary by Positions')
```

## In the Same Position, What does the Average Salary Different by Seniority?

```python
pt2 = pd.pivot_table(df, index=['job_title', 'seniority'], values='salary_avg').unstack()
pt2.sort_values('job_title', ascending=False, inplace=True)
pt2.columns.names = ([None, None])
pt2.columns = pt2.columns.droplevel()

pt2.plot(kind='barh', stacked=True)
plt.xlabel('salary_avg')
adjust_plot(title='Average Salary by Positions and Seniority')
```

## What is the Average Salary in Different Sates?

```python
pt3 = pd.pivot_table(df, index='location', values='salary_avg').sort_values('salary_avg', ascending=False)

sns.barplot(pt3, x='salary_avg', y=pt3.index, orient='h')
adjust_plot('Average Salary by States')
```

## What is Data Analyst's Average Salary in Different States?

```python
pt4 = pd.pivot_table(df[df['job_title'] == 'data analyst'], index='location', values='salary_avg')
pt4.sort_values('salary_avg', ascending=False, inplace=True)

sns.barplot(pt4, x='salary_avg', y=pt4.index, orient='h')
adjust_plot("Data Analyst's Average Salary by States")
```

## What is Data Scientist's Average Salary in Different States?

```python
pt5 = pd.pivot_table(df[df['job_title'] == 'data scientist'], index='location', values='salary_avg')
pt5.sort_values('salary_avg', ascending=False, inplace=True)

sns.barplot(pt5, x='salary_avg', y=pt5.index, orient='h')
adjust_plot("Data Scientist's Average Salary by States")
```

## What does the Average Salary Different by Position in Different States?

```python
pt6 = pd.pivot_table(df, index=['location', 'job_title'], values='salary_avg').unstack()
pt6.sort_values('location', ascending=False, inplace=True)
pt6.columns.names = ([None, None])
pt6.columns = pt6.columns.droplevel()

pt6.plot(kind='barh', stacked=True)
adjust_plot(title='Average Salary by States and Positions')
```

## What does the Average Salary Different by Seniority in Different States

```python
pt7 = pd.pivot_table(df, index=['location', 'seniority'], values='salary_avg').unstack()
pt7.sort_values('location', ascending=False, inplace=True)
pt7.columns.names = ([None, None])
pt7.columns = pt7.columns.droplevel()

pt7.plot(kind='barh', stacked=True)
plt.xlabel('')
adjust_plot(title='Average Salary by States and Seniority')
```

## How many jobs are there available in Different States?

```python
pt8 = pd.pivot_table(df, index=['location', 'seniority'], values='salary_avg', aggfunc='count').unstack()
pt8.sort_values('location', ascending=False, inplace=True)
pt8.columns.names = ([None, None])
pt8.columns = pt8.columns.droplevel()

pt8.plot(kind='barh', stacked=True, sort_columns=True)
adjust_plot(title='Number of Jobs by States and Seniority')
```

## What do Companies Mention in their Job Descriptions?

```python
text = ' '.join(df['job_description'])

# remove stop words and punctuationa
stop = set(stopwords.words('english'))
tokens = word_tokenize(text)

new_text = []
for token in tokens:
    if token not in stop and token.isalpha():
        new_text.append(token.lower())
new_text = ' '.join([token for token in new_text])

# draw the word cloud
wc = WordCloud(background_color='white', random_state=1, max_words=2000, width=1500, height=800)
wc.generate(new_text)

plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()       
```
